```python
import random


class Account:
    # Q2: 생성된 계좌 수 저장 (클래스 변수)
    account_count = 0

    def __init__(self, owner, balance):
        # Q1
        self.bank = "SC은행"
        self.owner = owner
        self.balance = int(balance)

        # 계좌번호: 3자리-2자리-6자리 랜덤 생성
        self.account_number = (
            f"{random.randint(0, 999):03d}-"
            f"{random.randint(0, 99):02d}-"
            f"{random.randint(0, 999999):06d}"
        )

        # Q7: 입금 횟수
        self.deposit_count = 0

        # Q10: 거래 내역
        self._deposit_logs = []    # [("입금"/"이자", amount), ...]
        self._withdraw_logs = []   # [("출금", amount), ...]

        Account.account_count += 1

    # Q3: 클래스 변수 출력 메서드
    @classmethod
    def get_account_num(cls):
        return cls.account_count

    # Q4: 입금(최소 1원)
    def deposit(self, amount):
        amount = int(amount)
        if amount < 1:
            return False

        self.balance += amount
        self.deposit_count += 1
        self._deposit_logs.append(("입금", amount))

        # Q7: 5회마다 1% 이자
        if self.deposit_count % 5 == 0:
            interest = int(self.balance * 0.01)
            self.balance += interest
            self._deposit_logs.append(("이자", interest))

        return True

    # Q5: 출금(잔고 이상 불가)
    def withdraw(self, amount):
        amount = int(amount)
        if amount < 1:
            return False
        if amount > self.balance:
            return False

        self.balance -= amount
        self._withdraw_logs.append(("출금", amount))
        return True

    # Q6: 정보 출력(쉼표)
    def display_info(self):
        print(
            f"은행이름: {self.bank}, "
            f"예금주: {self.owner}, "
            f"계좌번호: {self.account_number}, "
            f"잔고: {self.balance:,}원"
        )

    # Q10: 입금/출금 내역 출력
    def deposit_history(self):
        for kind, amt in self._deposit_logs:
            print(f"{kind}: {amt:,}원")

    def withdraw_history(self):
        for kind, amt in self._withdraw_logs:
            print(f"{kind}: {amt:,}원")


# ============================
# 여기부터 “문제 1~10 검증 출력”
# ============================

print("Q1) Account 클래스 생성자")
acc1 = Account("파이썬", 10000)
print("요구사항: 은행이름=SC은행, 예금주/잔액 설정, 계좌번호 3-2-6 형태 랜덤 생성")
print("결과 확인:")
acc1.display_info()
print()

print("Q2) 클래스 변수로 생성된 계좌 수 저장")
print("요구사항: Account로 계좌가 만들어질 때마다 count 증가")
print("결과 확인: 현재 생성된 계좌 수 =", Account.account_count)
print()

print("Q3) get_account_num()으로 생성된 계좌 수 출력")
print("요구사항: Account.get_account_num()이 계좌 수를 반환")
print("결과 확인: Account.get_account_num() =", Account.get_account_num())
print()

print("Q4) deposit() 입금 메서드 (최소 1원 이상)")
print("요구사항: 0원/음수 입금은 실패(False), 1원 이상 성공(True)")
print("결과 확인:")
print(" - deposit(0)  ->", acc1.deposit(0))
print(" - deposit(1)  ->", acc1.deposit(1))
print(" - 입금 후 잔고 ->", f"{acc1.balance:,}원")
print()

print("Q5) withdraw() 출금 메서드 (잔고 이상 출금 불가)")
print("요구사항: 잔고보다 큰 금액 출금 실패(False), 가능하면 성공(True)")
print("결과 확인:")
too_much = acc1.balance + 1
print(f" - withdraw({too_much:,}) ->", acc1.withdraw(too_much))
print(" - withdraw(1) ->", acc1.withdraw(1))
print(" - 출금 후 잔고 ->", f"{acc1.balance:,}원")
print()

print("Q6) display_info() 정보 출력 (잔고 쉼표 포함)")
print("요구사항: (잔고: 10,000원)처럼 3자리마다 쉼표")
print("결과 확인:")
acc1.display_info()
print()

print("Q7) 입금 5회마다 1% 이자 지급")
print("요구사항: 입금 횟수가 5회가 될 때 잔고의 1%가 잔고에 추가")
acc7 = Account("이자테스트", 10000)
print("테스트 준비: 초기 잔고 10,000원 계좌 생성")
print("입금 1,000원을 5회 수행 (5회째에 이자 지급)")
for i in range(5):
    acc7.deposit(1000)
print("결과 확인: 잔고 =", f"{acc7.balance:,}원")
print(" - 기대 흐름: 10,000 + 1,000*5 = 15,000 / 이자 1% = 150 / 최종 15,150")
print(" - 입금 내역(이자 포함):")
acc7.deposit_history()
print()

print("Q8) Account 인스턴스 3개 이상 생성 후 리스트에 저장")
accounts = [
    Account("A", 500000),
    Account("B", 1500000),
    Account("C", 2000000),
]
print("요구사항: 3개 이상 생성해서 리스트에 저장")
print("결과 확인: 리스트 길이 =", len(accounts))
print()

print("Q9) 리스트 순회: 잔고 100만원 이상 고객만 출력")
print("요구사항: balance >= 1,000,000인 계좌만 display_info() 출력")
print("결과 확인:")
for acc in accounts:
    if acc.balance >= 1_000_000:
        acc.display_info()
print()

print("Q10) 입금/출금 내역 기록 + history 메서드")
print("요구사항: deposit/withdraw 내역이 기록되고 출력 가능")
acc10 = Account("내역테스트", 10000)
acc10.deposit(5000)
acc10.withdraw(3000)
acc10.deposit(2000)
print("결과 확인: 현재 정보")
acc10.display_info()
print(" - deposit_history():")
acc10.deposit_history()
print(" - withdraw_history():")
acc10.withdraw_history()
```

## Quest 2 간단한 자동사냥 RPG 게임 만들기

```python
from __future__ import annotations
from dataclasses import dataclass
from random import randint
from typing import Dict


# ---------------------------
# Q1. Character 클래스
# ---------------------------
@dataclass
class Character:
    name: str
    level: int
    hp: int
    attack: int
    defense: int

    def is_alive(self) -> bool:
        """현재 체력이 0 이상인지 반환"""
        return self.hp >= 0

    def take_damage(self, damage: int) -> int:
        """
        공격을 받았을 때 (damage - defense) 만큼 체력 감소
        - 방어력이 데미지보다 크면 체력 감소 없음
        반환: 실제로 감소한 데미지(0 이상)
        """
        real_damage = max(0, damage - self.defense)
        self.hp -= real_damage
        return real_damage

    def attack_target(self, target: "Character") -> int:
        """
        타겟에게 데미지를 입힘
        - 데미지는 1~공격력 사이 랜덤 정수
        반환: 실제로 타겟에게 들어간 데미지(방어력 반영 후)
        """
        damage = randint(1, self.attack)
        real_damage = target.take_damage(damage)
        return real_damage


# ---------------------------
# Q2. Player / Monster 클래스
# ---------------------------
class Player(Character):
    def __init__(self, name: str):
        super().__init__(name=name, level=1, hp=100, attack=25, defense=5)
        self.exp = 0  # 경험치 속성

    def gain_experience(self, amount: int) -> None:
        """인수로 받은 정수만큼 경험치 획득"""
        self.exp += amount

    def level_up(self) -> None:
        """
        현재 경험치가 500 이상이면
        레벨 +1, 공격력 +10, 방어력 +5
        (요구사항 기준: 500 이상일 때 1회 상승)
        """
        if self.exp >= 500:
            self.level += 1
            self.attack += 10
            self.defense += 5


class Monster(Character):
    def __init__(self, name: str, level: int):
        hp = randint(10, 30) * level
        attack = randint(5, 20) * level
        defense = randint(1, 5) * level
        super().__init__(name=name, level=level, hp=hp, attack=attack, defense=defense)


# ---------------------------
# Q3. battle 함수
# ---------------------------
def battle(player: Player, monster: Monster) -> bool:
    """
    둘 중 하나의 체력이 0 이하가 될 때까지 공격을 주고받는 함수
    - 플레이어 생존 시: 경험치 (몬스터레벨*20) 추가, 레벨업 호출, '전투 승리!' 출력
    - 플레이어 사망 시: '전투 패배..' 출력
    반환: 플레이어가 살아남았는지 여부
    """
    print(f"\n{monster.name}과의 전투를 시작합니다.")

    while player.hp > 0 and monster.hp > 0:
        # Player -> Monster
        dmg = player.attack_target(monster)
        print(f"{player.name}이(가) {monster.name}에게 {dmg} 만큼 공격했다...!")
        print(f"{monster.name}의 체력: {monster.hp}")

        if monster.hp <= 0:
            break

        # Monster -> Player
        dmg = monster.attack_target(player)
        print(f"{monster.name}이(가) {player.name}에게 {dmg} 만큼 공격했다...!")
        print(f"{player.name}의 체력: {player.hp}")

    if player.hp > 0:
        player.gain_experience(monster.level * 20)
        player.level_up()
        print("전투 승리!")
        return True
    else:
        print("전투 패배..")
        return False


# ---------------------------
# Q4. main 함수
# ---------------------------
def main() -> None:
    monster_dict: Dict[str, int] = {"슬라임": 1, "고블린": 2, "오크": 3}

    user_name = input("사용자 이름을 입력하세요: ").strip()
    if not user_name:
        user_name = "플레이어"

    player = Player(user_name)

    # 슬라임 -> 고블린 -> 오크 순서로 자동사냥
    for m_name, m_level in monster_dict.items():
        monster = Monster(m_name, m_level)
        alive = battle(player, monster)

        if not alive:
            print("게임오버")
            return

    print("\n모든 몬스터를 처치했습니다! 게임 클리어!")


if __name__ == "__main__":
    main()
```
