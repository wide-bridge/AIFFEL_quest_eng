Q10. 입금과 출금 내역이 기록되도록 코드를 업데이트
(입금 내역과 출금 내역을 출력하는 deposit_history와 withdraw_history 메서드 추가

```python
from random import randint

class Account:
    # Q 2 클래스 변수
    account_cnt = 0
    def __init__(self, name, balance):
        self.deposit_cnt = 0
        self.deposit_log = []
        self.withdraw_log = []
        
        self.name = name
        self.balance = int(balance)
        self.bank = 'SC은행'

        nums = [3, 2, 6]
        parts = []

        for n in nums:
            temp = ''
            for i in range(n):
                temp += str(randint(0, 9))
            parts.append(temp)

        self.account_number ='-'.join(parts)
        # Q 2 계좌개수 증가
        Account.account_cnt += 1
        
    # Q3 Account 클래스로부터 생성된 계좌의 개수를 출력하는 get_account_num() 메서드 추가
    @classmethod
    def get_account_num(cls):
        print(cls.account_cnt)
        
    # Q4. 입금메서드: Account 클래스에 입금을 위한 deposit 메서드 추가. 입금은 최소 1원 이상만 가능
    # Q7. 이자지급하기: 입금횟수가 5회가 될 때 잔고를 기준으로 1%의 이자가 잔고에 추가 되도록 코드 변경
    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount
            self.deposit_log.append(amount) # Q 10

            self.deposit_cnt += 1
            if self.deposit_cnt % 5 == 0:
                # 이자 지급
                self.balance = self.balance * 1.01
        else:
            print('입금은 1원 이상만 가능합니다.')
            
    # Q5. 출금 메서드: Accout 클래스에 출금을 위한 withdraw 메서드 추가, 출금은 계좌의 잔고 이상 출금 불가
    def withdraw(self, amount):
        if self.balance > amount:
            self.balance -= amount
            self.withdraw_log.append(amount) # Q 10 
            return f'{amount}원이 인출되었습니다. 잔고는 {self.balance}원 입니다.'
        else:
            return f'인출하고자 하는 금액({amount}원)이 잔고({self.balance}원)를 초과하여 인출할 수 없습니다.'

    # Q6. 정보출력 메서드: Account 인스턴스에 저장된 정보를 출력하는 display_info() 메서드 추가. 잔고는 세자리마다 쉼표를 출력
    # (은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고 10,000원)
    def display_info(self):
        print("은행이름:", self.bank)
        print("예금주:", self.name)
        print("계좌번호:", self.account_number)
        print('잔고:', f'{self.balance:,}원')

    # Q10. 입금과 출금 내역이 기록되도록 코드를 업데이트
    #(입금 내역과 출금 내역을 출력하는 deposit_history와 withdraw_history 메서드 추가
    def deposit_history(self):
        for i in self.deposit_log:
            print(i)

    def withdraw_history(self):
        for i in self.withdraw_log:
            print(i)

k = Account("Kim", 1000)
k.deposit(100)
k.deposit(200)
k.deposit(300)
k.deposit_history()

print('-' * 50)

k.withdraw(100)
k.withdraw(200)
k.withdraw_history()
```

```python
from __future__ import annotations
from dataclasses import dataclass
from random import randint
from typing import Dict


# ---------------------------
# Q1. Character 클래스
# ---------------------------
@dataclass
class Character:
    name: str
    level: int
    hp: int
    attack: int
    defense: int

    def is_alive(self) -> bool:
        """현재 체력이 0 이상인지 반환"""
        return self.hp >= 0

    def take_damage(self, damage: int) -> int:
        """
        공격을 받았을 때 (damage - defense) 만큼 체력 감소
        - 방어력이 데미지보다 크면 체력 감소 없음
        반환: 실제로 감소한 데미지(0 이상)
        """
        real_damage = max(0, damage - self.defense)
        self.hp -= real_damage
        return real_damage

    def attack_target(self, target: "Character") -> int:
        """
        타겟에게 데미지를 입힘
        - 데미지는 1~공격력 사이 랜덤 정수
        반환: 실제로 타겟에게 들어간 데미지(방어력 반영 후)
        """
        damage = randint(1, self.attack)
        real_damage = target.take_damage(damage)
        return real_damage


# ---------------------------
# Q2. Player / Monster 클래스
# ---------------------------
class Player(Character):
    def __init__(self, name: str):
        super().__init__(name=name, level=1, hp=100, attack=25, defense=5)
        self.exp = 0  # 경험치 속성

    def gain_experience(self, amount: int) -> None:
        """인수로 받은 정수만큼 경험치 획득"""
        self.exp += amount

    def level_up(self) -> None:
        """
        현재 경험치가 500 이상이면
        레벨 +1, 공격력 +10, 방어력 +5
        (요구사항 기준: 500 이상일 때 1회 상승)
        """
        if self.exp >= 500:
            self.level += 1
            self.attack += 10
            self.defense += 5


class Monster(Character):
    def __init__(self, name: str, level: int):
        hp = randint(10, 30) * level
        attack = randint(5, 20) * level
        defense = randint(1, 5) * level
        super().__init__(name=name, level=level, hp=hp, attack=attack, defense=defense)


# ---------------------------
# Q3. battle 함수
# ---------------------------
def battle(player: Player, monster: Monster) -> bool:
    """
    둘 중 하나의 체력이 0 이하가 될 때까지 공격을 주고받는 함수
    - 플레이어 생존 시: 경험치 (몬스터레벨*20) 추가, 레벨업 호출, '전투 승리!' 출력
    - 플레이어 사망 시: '전투 패배..' 출력
    반환: 플레이어가 살아남았는지 여부
    """
    print(f"\n{monster.name}과의 전투를 시작합니다.")

    while player.hp > 0 and monster.hp > 0:
        # Player -> Monster
        dmg = player.attack_target(monster)
        print(f"{player.name}이(가) {monster.name}에게 {dmg} 만큼 공격했다...!")
        print(f"{monster.name}의 체력: {monster.hp}")

        if monster.hp <= 0:
            break

        # Monster -> Player
        dmg = monster.attack_target(player)
        print(f"{monster.name}이(가) {player.name}에게 {dmg} 만큼 공격했다...!")
        print(f"{player.name}의 체력: {player.hp}")

    if player.hp > 0:
        player.gain_experience(monster.level * 20)
        player.level_up()
        print("전투 승리!")
        return True
    else:
        print("전투 패배..")
        return False


# ---------------------------
# Q4. main 함수
# ---------------------------
def main() -> None:
    monster_dict: Dict[str, int] = {"슬라임": 1, "고블린": 2, "오크": 3}

    user_name = input("사용자 이름을 입력하세요: ").strip()
    if not user_name:
        user_name = "플레이어"

    player = Player(user_name)

    # 슬라임 -> 고블린 -> 오크 순서로 자동사냥
    for m_name, m_level in monster_dict.items():
        monster = Monster(m_name, m_level)
        alive = battle(player, monster)

        if not alive:
            print("게임오버")
            return

    print("\n모든 몬스터를 처치했습니다! 게임 클리어!")


if __name__ == "__main__":
    main()
```
